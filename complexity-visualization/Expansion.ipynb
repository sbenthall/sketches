```python
import evarr

import matplotlib.pyplot as plt
import numpy as np
```

```python
it = 10
parts = .455

a = np.random.random((2,2))

for i in range(it):
    print(i)
    o = a
    a = evarr.x2(a)
    n = np.random.random(a.shape)
    
    r0 = np.random.randint(0, a.shape[0])
    r1 = np.random.randint(0, a.shape[1])
    
    a = np.round(evarr.zoom(a))
    a = (parts * a + n) / (parts + 1)
    a = np.roll(a, (r0,r1))
    
```

```python
plt.figure(figsize=(9,9))
plt.imshow(np.round(a))
```
