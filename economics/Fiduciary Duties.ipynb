### Fiduciary Duties

In business law, many people have affirmative obligations and prohibitions owards each other. Examples include the relationship between the management of a business and its shareholders, the relationship of a financial advisor to a customer, or the relationship of a lawyer to a client. Discussion of fiduciary duties has been vitalized by Balkin (2016), who has proposed that information fiduciary duties may be a solution to the misgivings about the consumer Internet and personal data use.

Fiduciary duties are an area of unsettled law whose interpretation varies with jurisdiction. However, they are not arbitrary, and progress towards understanding the content of the duties can be made through doctrinal research. The duties of Care, Loyalty, Good Faith, Prudence, Confidentiality, and Disclosure reappear across many sectors of law and are often explicit about the obligations and prohibitions implied by them.

In our investigation of fiduciary duties, we find that many of them have structure similar to the transmission principles of information norms in Contextual Integrity (Nissenbaum, 2009). They often refer to the conditions under which an action may be taken; often the action, or the condition, are defined in terms of the flow of information. Following Benthall (2018, 2019), we will model these requirements in terms of _situated information flows_, causal flows situated in a causal context, with multiple agents modeled using Koller and Milch's (2003) _multi-agent influence diagram_ framework. Our work will aim to clarify the meaning of fiduciary obligations with a precise interpretation suitable for applications in economics and computer science.


### Modeling

This notebook will model fiducariary duties as Multi-Agent Influence Diagrams (Koller and Milch, 2003). Following Benthall (2018), I will adopt the following notation:
 * Unmarked capital letters $A, B, C$ for chance nodes
 * Tilded capital letters $\tilde{A}, \tilde{B}, \tilde{C}$ for strategic nodes
 * Breved capital letters $\breve{A}, \breve{B}, \breve{C}$ for utility nodes
 * A hatted capital letter $\hat{A}, \hat{B}, \hat{C}$ is a strategic node that has been assigned conditional probability distribution via a strategically optimal decision rule.
 * Subscript for the 'player' who controls the strategic or utility variables, $\tilde{A}_q, \breve{A}_q$
 
For graphics, we will adopt the original Koller and Milch notation of elipses for chance variables, rectangles for strategic variables, and rhombuses for utility variables.

For the sake of clarity, we will consistently refer to the principal $b$ (Bob) and the agent $a$ (Alice); the latter has the duty to the former. We will include another player $e$ (Eve) as the third party who may or may not be trying to incentivize Alice to betray her duties to Bob.

## Duty of Confidentiality

A negative rule about information flow. The agent may not disclose information (e.g about the principal) for their own benefit.

**Example**. Bob tells some secret information $S$ to Alice, who also learns for some chance/contextual information $C$. She compiles a report and delivers it to Eve $E$. Eve is trying to learn Bob's secret and does so by offering an incentive to Alice to include this information in her report. Alice would respond to this incentive and do so unless prohibited by a fiduciary duty.

The fiduciary duty applies to the decision rule $\delta_R$ corresponding to the strategic action of writing the report $R$ and its informativeness about the secret $S$. It requires that the result of the action does not contain information about the secret. Let $I(X;Y)$ be the mutual information between two random variables $X$ and $Y$. Then the report maintains confidentiality about the secret if:

$$I(\delta_R(\hat{S}_b,C);\hat{S}_b) = 0$$
